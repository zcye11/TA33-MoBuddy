```python
import pandas as pd
import geopandas as gpd
```

```python
gdf = gpd.read_file('./FootpathGrades.geojson')
```

```python
gdf
```

```python
gdf.plot(figsize=(18, 18), alpha = 1)
```

```python
gdf.info()
```

```python
gdf['grade1in'].head()
```

```python
green = gdf[gdf.grade1in.astype('float') <= 20].assign(color='green')
yellow = gdf[(gdf.grade1in.astype('float') > 20) & (gdf.grade1in.astype('float') <= 50)].assign(color='yellow')
red = gdf[gdf.grade1in.astype('float') > 50].assign(color='red')
```

```python
mel = green.append(yellow)
mel = mel.append(red)
```

```python
green.info()
```

```python
yellow.info()
```

```python
red.info()
```

```python
mel
```

```python
mel.plot(figsize=(18, 18), alpha = 1, color=mel['color']) 
```

```python
mel.to_file("mel_footpath.geojson", driver='GeoJSON')
```
